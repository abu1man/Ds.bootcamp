```python
# importing the libraries for the algorithm
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
# for building the model 
from sklearn.model_selection import train_test_split
# for preprocessing 
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler # fro normalizing the data
# for evalution 
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report # it include all the metices
```

```python
# Prepocesing the data
encoder = LabelEncoder() #initialise the encoder
# my_data=data  for converting string to int or the below way 

data.Gender = encoder.fit_transform(data.Gender)
X = data[['User ID', 'Gender', 'Age', 'EstimatedSalary',]]
y= data["Purchased"]
encoder = encoder.fit_transform(y)

# # display y
data
encoder
```

```python
X
```

```python
# Splitting the dataset
X_train,X_test,y_train,y_test = train_test_split( X,y, test_size = 0.3, random_state= 42)

# display the size
X_train.shape,X_test.shape, y_test.shape, y_train.shape
```

```python
# Normalisation
Scaler = StandardScaler()
X_train = Scaler.fit_transform(X_train)
X_test = Scaler.fit_transform(X_test)
```

```python
# Building the model
decision_model = DecisionTreeClassifier()  #initializing the models
decision_model.fit(X_train,y_train)  #model fitting

# Prediction
y_predict_decision = decision_model.predict(X_test) #making prediction
```

```python
# Evaluation
accuracy = accuracy_score(y_test, y_predict_decision)
accuracy
```

```python
# Initialising all the models
random_model = RandomForestClassifier()
kn_model = KNeighborsClassifier()


# Fitting the models
random_model.fit(X_train, y_train)
kn_model.fit(X_train, y_train)


# Predicting the models
y_predict_random = random_model.predict(X_test)
y_predict_kn = kn_model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_predict_decision)
print("accuracy")
print(accuracy)
print("**********************************************************************")


random_accuracy = accuracy_score(y_test, y_predict_random)
print("random_accuracy:")
print(random_accuracy)
print("**********************************************************************")


kn_accuracy = accuracy_score(y_test, y_predict_kn)
print( "kn_accuracy:")
print( kn_accuracy)
print("**************************************************************************")
```
