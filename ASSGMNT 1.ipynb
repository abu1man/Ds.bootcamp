```python
snb.boxplot(x="sepal_length", y="sepal_width", data= data)
plt.title("petal Distributuon of dataset")
plt.show()
```

```python
# scattered-plot
# sepal_length VS sepal_width
snb.scatterplot(x="sepal_width", y="sepal_length", hue="species", data = data)
plt.title("A scatterPlot for the three different Species")
# Draw an arrow to the data point with the maximum sepal length
plt.arrow(data["sepal_length"].max(), data["sepal_width"].max(), -2, 2,
          head_width=0.4, head_length=0.3, color="red")
plt.show()
```

```python
# scattered-plot
# petal_length VS petal_width
snb.scatterplot(x="petal_length", y="petal_width", hue="species", data = data)
plt.title("A scatterPlot for the three different Species")
# Draw an arrow to the data point with the maximum sepal length
plt.arrow(data["sepal_length"].max(), data["sepal_width"].max(), -2, 2,
          head_width=0.5, head_length=0.5, color="red")
plt.show()
```
